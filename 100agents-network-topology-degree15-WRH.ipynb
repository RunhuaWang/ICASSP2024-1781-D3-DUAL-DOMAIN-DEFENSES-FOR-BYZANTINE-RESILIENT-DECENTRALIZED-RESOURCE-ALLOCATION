# 问题介绍：生成100个用户，每个用户度均为15的网络拓扑对应的邻接矩阵

```python
import numpy as np
import random
```

```python
Adjacency_Matrix  = np.zeros(shape=(100,100))#定义100*100维的邻接矩阵
sum_list=[]#共100个数据
connect_number=[]#100个用户被连接的次数
```

```python
count=0
while sum_list.count(15)!=200:
    count=count+1
    print("第",(count),"次循环")
    sum_list.clear()
    break_flag=False
    Adjacency_Matrix  = np.zeros(shape=(100,100))
    for i in range(0,100):
        if (i==0):
            Adjacency_Matrix[i][i+1]=1
            Adjacency_Matrix[i][99]=1
        elif (i==99):
            Adjacency_Matrix[i][0]=1
            Adjacency_Matrix[i][i-1]=1
        else:
            Adjacency_Matrix[i][i+1]=1
            Adjacency_Matrix[i][i-1]=1
            
    connect_number.clear()#100个用户被连接的次数
    for i in range (0,100):
        connect_number.append(2)
    
    
    #开始为邻接矩阵赋值
    for i in range (0,100):
        random_code=[]
        k=15
        list=[]
        for j in range(0,100):
            if (Adjacency_Matrix[i][j]==1):
                k=k-1
                list.append(j)
        print("第",count,"循环中第",i,"个用户的k值为：",k)
        print("第",count,"循环中第",i,"个用户的list值为：",list)
        
        if (k<=0):
            continue            
        if (k>0):
            number1=0
            number2=0
            for j in range(i+1,100):
                if (connect_number[j]>=15):
                    number1=number1+1
                if (Adjacency_Matrix[i][j]==1):
                    number2=number2+1
                if (connect_number[j]>=15 and Adjacency_Matrix[i][j]==1):
                    number1=number1-1                      
            if (99-i-number1-number2<k):
                break_flag=True
            else:
                e=0
                q=0
                while e!=k or q!=len(list)*k:
                    e=0
                    q=0
                    random_code=[]
                    random_code=random.sample(range(i+1, 100), k)
                    for j in range(0,k):
                        if (connect_number[random_code[j]]<15):
                            e=e+1
                    for p in range(0,len(list)):
                        for j in range(0,k):
                            if (list[p]!=random_code[j]):
                                q=q+1
                for j in range(0,k):
                    Adjacency_Matrix[i][random_code[j]]=1
                    Adjacency_Matrix[random_code[j]][i]=1
                    connect_number[random_code[j]]=connect_number[random_code[j]]+1
                connect_number[i]=15
                print("第",count,"循环中第",i,"个用户的random_code值为：",random_code)
                print("第",count,"循环中第",i,"个用户的connect_number值为：",connect_number)
            if (break_flag==True):
                continue
            
    print("第",count,"次循环的邻接矩阵",Adjacency_Matrix)
    
    print("行和")
    for i in range(0,100):
        sum=0
        for j in range(0,100):
            sum=sum+Adjacency_Matrix[i][j]
        sum_list.append(sum)
        print(sum_list[i])
    print("*********************************")
    print("列和")
    for i in range(0,100):
        sum=0
        for j in range(0,100):
            sum=sum+Adjacency_Matrix[j][i]
        sum_list.append(sum)
        print(sum_list[i])
    print("***********************************************************************")   
```

```python
Double_Random_Adjacency_Matrix=Adjacency_Matrix
```

```python
#将邻接矩阵转化成双随机的邻接矩阵
for i in range(0,100):
    for j in range(0,100):
        if (Double_Random_Adjacency_Matrix[i][j]==1):
            Double_Random_Adjacency_Matrix[i][j]=1/15  
```

```python
np.savetxt('100_Network_Topology_Double_Random_Adjacency_Matrix_Degree_15.txt',Double_Random_Adjacency_Matrix,fmt='%0.8f')#数据存储
```
